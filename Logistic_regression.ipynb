```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("diabetes2.csv")
df.head()
```

```python
x=df.drop(columns="Outcome")
y=df["Outcome"]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a scatter plot matrix of the features
sns.pairplot(df, hue='Outcome')
plt.suptitle("Scatter Plot Matrix of Features by Outcome", y=1.02)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=23)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
x_train_scaled = scaler.fit_transform(X_train)
x_test_scaled = scaler.transform(X_test)
```

```python
lr=LogisticRegression(max_iter=50)
lr.fit(x_train_scaled,y_train)
```

```python
acc=accuracy_score(y_test,lr.predict(x_test_scaled))
print(acc)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,lr.predict(x_test_scaled));
```

```python
print(classification_report(y_test,lr.predict(x_test_scaled)));
```
